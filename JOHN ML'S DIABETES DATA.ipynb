### Visualisation was done to know the distribution of the Numerical features so as to choose the correct Scaler. And here, MinMax Scaler was choosen.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

features["scaled_age"] = scaler.fit_transform(features[["age"]])
features["scaled_bmi"] = scaler.fit_transform(features[["bmi"]])
features["scaled_HbA1c_level"] = scaler.fit_transform(features[["HbA1c_level"]])
features["scaled_blood_glucose_level"] = scaler.fit_transform(features[["blood_glucose_level"]])


features.drop(['age','bmi','HbA1c_level','blood_glucose_level'], axis=1, inplace=True)
```

## Machine Learning

```python
### Data Splitting.

from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest = train_test_split(features,label,test_size=0.2,random_state=42)
```

```python
## Model Training

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=42)

## Model Fitting/Training

model.fit(Xtrain,ytrain)
```

```python
## Model Prediction
predictions=model.predict(Xtest)

(predictions)
```

```python
ytest.head(5)
```

```python
##Evaluation of the model

from sklearn.metrics import classification_report,confusion_matrix
report = classification_report(ytest,predictions)
```

```python
print(report)
```

```python
matrix = confusion_matrix(ytest,predictions)

matrix
```

```python
## Visualise the confusion matrix

sns.heatmap(matrix, annot = True, fmt = 'g')
plt.xlabel('Prediction')
plt.ylabel('Actual')
```

Explaining the confusion matrix.

1.TRUE NEGATIVE (17317) - This means the model predicted 0 when the answer is 0 as well.

3.FALSE NEGATIVE (655) - This means the model predicted 0 when the actual answer is 1.

4.TRUE POSITIVE (1066) - This means the model predicted 1 when the answe is actually 1 as well.

```python
### We are now checking to improve this model by applying class weight.


model = LogisticRegression(random_state=42, class_weight={0:1,1:4})
model.fit(Xtrain,ytrain)
predictions = model.predict(Xtest)
report = classification_report(ytest,predictions, digits=5)
print(report)
matrix = confusion_matrix(ytest,predictions)
sns.heatmap(matrix, annot = True, fmt = 'g')
plt.xlabel('prediction')
plt.ylabel('Actual Value')
plt.show()
```

```python
### We are going to check other classification models now



from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier, AdaBoostClassifier

from xgboost import XGBClassifier


class_weights = {0:1, 1:4}


classifiers = [[SGDClassifier(random_state = 42, class_weight = class_weights), 'SGD Classifier'],
[KNeighborsClassifier(), 'KNeighbors Classifiers'],
[DecisionTreeClassifier(random_state = 42, class_weight = class_weights), 'Decision Tree'],
[RandomForestClassifier(random_state = 42, class_weight = class_weights), 'Random Forest'],
[GradientBoostingClassifier(random_state = 42), 'Gradient Boosting'],
[AdaBoostClassifier(random_state = 42), 'Ada Boost'],
[XGBClassifier(eval_metric = 'mlogloss'), 'XGBClassifier']]
    
   
    
    
for classifier in classifiers:
    model_name = classifier[1]
    model = classifier[0]
    print(f"--{model_name}--")
    model.fit(Xtrain,ytrain)
    predictions = model.predict(Xtest)
    report = classification_report(ytest,predictions)
    matrix = confusion_matrix(ytest,predictions)
    print("Classification report:\n", report)
    sns.heatmap(matrix, annot = True, fmt = 'g')
    plt.xlabel('predictions')
    plt.ylabel('Actual')
    plt.show()
 
    
    
    ###  print("Classification report:\n", report)  algorithm = "SAMME"
```

```python
from imblearn.over_sampling import SMOTE
```

```python
### Bringing back the features and label as X and y.

X = features.copy()
y = label.copy()
```

```python
### Apply SMOTE to oversample the minority class


### initialize the smote

smote = SMOTE(random_state = 42)

X_resampled, y_resampled = smote.fit_resample(X,y)


y_resampled.value_counts()
```

```python
## Do the visualization

sns.countplot(x=y_resampled)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_resampled,y_resampled,test_size=0.2,random_state=42)
```

```python
class_weights = "balanced"




classifiers = [[LogisticRegression(random_state = 42, class_weight = class_weights), 'Logistic Regression'],
[SGDClassifier(random_state = 42, class_weight = class_weights), 'SGD Classifier'],
[KNeighborsClassifier(), 'KNeighbors Classifiers'],
[DecisionTreeClassifier(random_state = 42, class_weight = class_weights), 'Decision Tree'],
[RandomForestClassifier(random_state = 42, class_weight = class_weights), 'Random Forest'],
[GradientBoostingClassifier(random_state = 42), 'Gradient Boosting'],
[AdaBoostClassifier(random_state = 42), 'Ada Boost'],
[XGBClassifier(eval_metric = 'mlogloss'), 'XGBClassifier']]
    
   
    
    
for classifier in classifiers:
    model_name = classifier[1]
    model = classifier[0]
    print(f"--{model_name}--")
    model.fit(X_train,y_train)
    predictions = model.predict(Xtest)
    report = classification_report(ytest,predictions)
    matrix = confusion_matrix(ytest,predictions)
    print("Classification report:\n", report)
    sns.heatmap(matrix, annot = True, fmt = 'g')
    plt.xlabel('predictions')
    plt.ylabel('Actual')
    plt.show()
 
```
